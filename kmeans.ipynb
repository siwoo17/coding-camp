```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# 데이터 만들기
사람들의 키와 몸무게 데이터를 만들어보도록 하겠습니다.

```python
df = pd.DataFrame(columns=['height', 'weight'])   #빈 데이터프레임 만들기
df.loc[0] = [185,60]                              #데이터프레임에 데이터 하나 추가
df.loc[1] = [180,60]
df.loc[2] = [185,70]
df.loc[3] = [165,63]
df.loc[4] = [155,68]
df.loc[5] = [170,75]
df.loc[6] = [175,80]
df.loc[7] = [172,67]
df.loc[8] = [177,73]
df.loc[9] = [169,62]
```

```python
df.head(10)     #기본 df.head()는 5개의 행을 보여줌, 그런데 ()안에 숫자를 써주면, 그 숫자만큼의 행을 보여줌
```

# 데이터 시각화
데이터를 차트에 시각화합니다.

```python
sns.lmplot('height', 'weight',     #seaborn 라이브러리 
           data=df, fit_reg=False, 
           scatter_kws={"s": 200, 'color':'red'})  #scatter_kws: 점을 다양하게 표현,s:크기
```

# k 평균 군집화
sklearn의 kmean 라이브러리에 데이터를 활용하여, 데이터를 군집화합니다.

```python
data_points = df.values   #군집화에 사용할 특징을 추출하여 변수에 저장, 변수에 저장되는 것은 array
kmeans = KMeans(n_clusters=2).fit(data_points)  #fit함수를 사용하여 모델을 만듦
#kmeans는 객체
```

아래 명령어로 각 군집의 중심 위치를 확인할 수 있습니다.

```python
kmeans.cluster_centers_  #k=3, 3개의 k의 (x,y), a,b,c
```

데이터가 어느 군집에 소속되어 있는 지 데이터프레임 cluster_id 행에 저장합니다.

```python
df['cluster_id'] = kmeans.labels_   #기존의 df에 열을 추가하여, 레이블 값을 추가한 열에 저장
```

데이터프레임을 조회하여, 데이터별 군집을 확인할 수 있습니다.

```python
df.head(10)
```

# k 평균 군집 시각화
군집 결과를 seaborn과 dataframe을 활용하여 손쉽게 시각화 할 수 있습니다.  
키가 작은 그룹, 중간 그룹, 큰 그룹으로 군집된 결과를 각기 다른 색으로 구분된 군집으로 확인 할 수 있습니다.

```python
sns.lmplot('height', 'weight', data=df, fit_reg=False,
           scatter_kws={"s": 150},
           hue="cluster_id")      #hue: 분류 카테고리(범례)
```
